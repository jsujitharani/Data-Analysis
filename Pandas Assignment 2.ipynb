```python
datac=data.copy()
```

```python
datac.fillna({"Title":"no title","Genre":"no genre","User Rating":0,"Number of Votes":0,"Runtime":"10","Year":0,"Summary":"no summary","Stars":"no stars","Certificate":"no certificate","Metascore":0,"Gross":10,"Episode":0,"Episode Title":"no"},inplace=True)
```

```python
datac.head(1)
```

```python
datac.replace({"User Rating":{"User Rating":0},"Number of Votes":{"Number of Votes":0},"Runtime":{"Runtime":"10"},"Metascore":{"Metascore":0},"Gross":{"Gross":10},"Episode":{"Episode":0}},inplace=True)
```

```python
datac=datac.astype({"User Rating":float,"Metascore":int,"Gross":int,"Episode":int,"Year":str})
```

```python
def func(x):
    return x.replace(" min","")
def f(x):
    return int(x)
def f1(x):
    return x.replace(",","")
datac["Runtime"]=datac["Runtime"].map(func).map(f1).map(f)
```

```python
def func(x):
    return x.replace(" min","")
def f(x):
    return int(x)
def f1(x):
    return x.replace(",","")
def f2(x):
    return str(x)
datac["Number of Votes"]=datac["Number of Votes"].map(f2).map(func).map(f1).map(f)
```
